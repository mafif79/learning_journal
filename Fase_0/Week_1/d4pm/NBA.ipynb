```python
#chaining + groupby
df[
    (df ['year_id']>2012)&
    (df['fran_id']=='Spurs')&
    (df['game_result']=='W')
].groupby('year_id')['pts'].mean()
```

```python
df[
    (df ['year_id']>2012)&
    (df['fran_id']=='Spurs')&
    (df['game_result']=='W')
].groupby(['year_id', 'game_location'])['pts'].mean()
```
